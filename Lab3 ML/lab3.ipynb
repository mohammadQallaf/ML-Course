Data loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("toyota_cars.xlsx")
df.head()
```

Data understanding

```python
df.shape
df.columns
df.info()
df.describe()
```

check missing values

```python
df.isnull().sum()
```

price destrebution

```python
plt.hist(df['price'])
plt.title("Price Distribution")
plt.show()
```

price boxplot

```python
sns.boxplot(x=df['price'])
plt.show()

```

**Bivariate Analysis**

Engine Displacement vs Price

```python
sns.scatterplot(x='engine_displacement', y='price', data=df)
plt.title("Engine Displacement vs Price")
plt.show()
```

Seats vs Price

```python
sns.boxplot(x='seats', y='price', data=df)
plt.title("Seats vs Price")
plt.show()
```

**Correlation Analysis**

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

## Conclusion

We explored the Toyota cars dataset to understand its structure and how the features relate to each other.

We found that cars with bigger engines usually have higher prices. The prices vary a lot between different car models, and there are some unusual values (outliers).

The correlation matrix showed how the numerical features are connected, which will help when building prediction models in the future.
